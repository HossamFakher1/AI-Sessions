```python
# Handling Outlier
fig, ax = plt.subplots(figsize=(50,16)) 
sns.boxplot(data=data, orient="v", palette="Set2")
plt.show()
```

```python
# Remove Outlier
def remove_outlier(col):
    sorted(col)
    Q1, Q3 = col.quantile([0.25, 0.75])
    IQR = Q3 - Q1
    print("Q1 = ", Q1, " Q3 = ", Q3, " IQR = ", IQR)
    lower_range = Q1 - (1.5 * IQR)
    upper_range = Q3 + (1.5 * IQR)
    return lower_range, upper_range

# Numerical data distribution
print(list(set(data.dtypes.tolist())))
df_num = data.select_dtypes(include = ["float64"])
lst_num_cols = df_num.columns.tolist()
#print("Numerical Data : \n",lst_num_cols)
print("\n")

indx = 0
for col in lst_num_cols:
    print(indx)
    lower_range, upper_range =  remove_outlier(data[col])
    data[col] = np.where(data[col] < lower_range, lower_range, data[col]) 
    data[col] = np.where(data[col] > upper_range, upper_range, data[col])
    indx = indx +1
    print("-----------------------------")
```

```python
# Handling Outlier
fig, ax = plt.subplots(figsize=(50,16)) 
sns.boxplot(data=data, orient="v", palette="Set2")
plt.show()
```

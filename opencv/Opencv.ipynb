```python
cap=cv2.VideoCapture(0)

cap.set(3,1280)
cap.set(4,720)

frame_width = int(cap.get(3))
frame_height = int(cap.get(4))

size = (frame_width, frame_height)


result = cv2.VideoWriter('filename.avi', 
                         cv2.VideoWriter_fourcc(*'MJPG'),
                         10, size)

while cap.isOpened() :
    ret,image=cap.read()
    if ret == False :
        break
    
    if cv2.waitKey(1)==ord('q'):
        break
    
    result.write(image)
    cv2.imshow('video',image)
    
cap.release()
result.release()
cv2.destroyAllWindows()
```

```python
def predict_image(img,MODEL,labels,size):
    img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    img=cv2.resize(img,(size,size))
    img=img/255.0
    img=np.expand_dims(img,axis=0)
    result=MODEL.predict(img)
    #result=np.round(result)[0]  #Sigmoid output 0.5 < result < 1.0         
    result = np.argmax(result,axis=-1)[0] # softmax output  sum ( result ) == 1 
    return labels[result]
```

```python
cap=cv2.VideoCapture(0)
model=tf.keras.models.load_model('model.hdf5')
while cap.isOpened() :
    ret,image=cap.read()
    if ret == False :
        break
    result=predict_image(img=image,MODEL=model,labels=['paper', 'rock', 'scissors'],size=224)
    image=cv2.putText(img=image, text=result, org=(100,100), fontFace=cv2.FONT_HERSHEY_SIMPLEX ,
                      fontScale=3, color=(0,0,255), thickness=2)
    if cv2.waitKey(1)==ord('q'):
        break
    cv2.imshow('video',image)
cap.release()
cv2.destroyAllWindows()
```
